# Test SageMaker Studio Connection
Verify AWS credentials and SageMaker access.

```python
import boto3
import sagemaker

# Use the AWS Toolkit profile with explicit region
session = boto3.Session(profile_name="YOUR_PROFILE", region_name="us-east-1")
print(f"Region: {session.region_name}")

sts = session.client("sts")
identity = sts.get_caller_identity()
print(f"Account ID: {identity['Account']}")
print(f"User ARN: {identity['Arn']}")

sm_client = session.client("sagemaker")
print(f"SageMaker endpoint: {sm_client.meta.endpoint_url}")
print("\nConnection successful!")
```
